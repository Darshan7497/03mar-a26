# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Some of the commonly used plots in Seaborn are:

Scatter plot: It is used to visualize the relationship between two continuous variables. Each point in the plot represents an observation in the dataset.

Line plot: It is used to visualize the trend of a continuous variable over time. It is commonly used in time-series analysis.

Bar plot: It is used to compare the values of a categorical variable across different groups. It is useful for visualizing the distribution of a variable across different categories.

Histogram: It is used to visualize the distribution of a continuous variable. It shows the number of observations that fall within a specified bin range.

Heatmap: It is used to visualize the correlation between two or more variables. It is useful for identifying patterns and relationships in the dataset.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
# Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", size="region", data=fmri)
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
# Note: pclass, age, and fare are columns in the titanic dataset.

```python
sns.load_dataset('titanic')
```

```python
titanic = sns.load_dataset("titanic")
sns.boxplot(x = 'pclass', y = 'age', data= titanic)
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
dmnd = sns.load_dataset('diamonds')
```

```python
dmnd
```

```python
sns.displot(data =dmnd, x ='price',hue = 'cut', kde=True)
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column  of the iris dataset.

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(data = iris, hue = 'species')
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_matrix = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights_matrix, cmap='Reds')
```
